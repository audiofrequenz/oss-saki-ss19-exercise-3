```python
#state (container,x-coord,y-coord,todo)
containers = [0,1,2,3] #0=empty, 1=white, 2=red, 3=blue
containerPositions = [1,2,3,4]
cranePositions = [0,1,2,3,4] #Position box
craneItems = [0,1,2,3]
toDos = [1,2,3,4,5,6] #1=store white, 2=restore white, 3=store red, 4=restore red, 5=store blue, 6=restore blue

states = []
for container in containers:
    for containerPos in containerPositions:
        for currPos in cranePositions:
            for craneItem in craneItems:
                for toDo in toDos:
                    state = (container, containerPos, currPos, craneItem, toDo)
                    states.append(state)
                
print(len(states), "should be -> '1536'")
#should be 1536
print(states)
```
